# Zadanie 3

```python
wine = pd.read_csv("wine_data.csv")
wine
```

```python
X = wine.iloc[:, 1:]
y = wine.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
best_score = 0
parameters = {}
```

```python
from tqdm import trange

for depth in range (1, 11):
    for depth in trange (1, 11):
        for min_split in range(2, 11):
             for min_leaf in range(1, 11):
                for criterion in ["gini", "entropy"]:
                    clf = DecisionTreeClassifier(max_depth = depth, min_samples_split = min_split, 
                                                     min_samples_leaf = min_leaf, criterion = criterion)
                    scores = cross_val_score(clf, X_train, y_train, cv = 5)
                    if scores.mean() > best_score:
                        parameters["best_depth"] = depth
                        parameters["best_min_split"] = min_split
                        parameters["best_min_leaf"] = min_leaf
                        parameters["best_criterion"] = criterion
                        best_score = scores.mean()
                        print("Accuracy: {}".format(scores.mean()))
                        print("max_depth: {}, min_samples_split: {}, min_samples_leaf: {}, criterion: {}".format(depth,
                                                                                                                 min_split,
                                                                                                                 min_leaf,
                                                                                                                 criterion))
```

```python
parameters
```

```python
clf_dtc = DecisionTreeClassifier(criterion = parameters["best_criterion"], max_depth = parameters['best_depth'], 
                                 min_samples_split = parameters["best_min_split"], 
                                 min_samples_leaf = parameters["best_min_leaf"])
```

```python
clf_dtc.fit(X = X_train, y = y_train)
```

```python
prediction_dtc = clf_dtc.predict(X_test)
```

```python
results = pd.DataFrame({"y_true":y_test, "y_dtc":prediction_dtc})
```

```python
results
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm_dtc = confusion_matrix(y_true = results["y_true"], y_pred = results["y_dtc"])
sns.heatmap(cm_dtc, annot = True, fmt="d")
cm = confusion_matrix(y_true = results["y_true"], y_pred = results["y_dtc"])
print(cm)
```

# Zadanie 4

```python
white_wine = pd.read_csv("winequality-white.csv", sep = ";")
white_wine
```

```python
X = white_wine.iloc[:, 0:10]
y = white_wine.iloc[:, -2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
best_score = np.inf
parameters = {}
```

```python
from sklearn.tree import DecisionTreeRegressor

for depth in range (1, 11):
    for depth in trange (1, 11):
        for min_split in range(2, 11):
            for min_leaf in range(1, 11):
                regressor = DecisionTreeRegressor(max_depth = depth,
                                                  min_samples_split = min_split,
                                                  min_samples_leaf = min_leaf)
                scores = cross_val_score(regressor, X_train, y_train, cv = 5)
                if scores.mean() < 0:
                    pass
                elif scores.mean() < best_score:
                    parameters["best_depth"] = depth
                    parameters["best_min_split"] = min_split
                    parameters["best_min_leaf"] = min_leaf
                    best_score = scores.mean()
```

```python
parameters
```

```python
reg_dtr = DecisionTreeRegressor(max_depth = parameters["best_depth"],
                                  min_samples_split = parameters["best_min_split"],
                                  min_samples_leaf = parameters["best_min_leaf"])
```

```python
reg_dtr.fit(X = X_train, y = y_train)
```

```python
prediction_dtr = reg_dtr.predict(X_test)
```

```python
results = pd.DataFrame({"y_true": y_test, "y_pred": prediction_dtr})
results
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_true = y_test, y_pred = prediction_dtr)
```
